```python
### (Urclass Quiz) Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer 

cancer_dataset = load_breast_cancer()

features = cancer_dataset.data
target = cancer_dataset.target
df = pd.DataFrame(data=features, columns=cancer_dataset.feature_names)
df['target'] = target

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=2)
print(X_train.shape, X_test.shape)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
xgbc = XGBClassifier().fit(X_train, y_train)
print(f1_score(y_test, xgbc.predict(X_test)))
```

### 3) Permutation importancesÎ•º Í≥ÑÏÇ∞ÌïòÍ≥† Îã§Î•∏ Ï§ëÏöîÎèÑ Í≤∞Í≥ºÏôÄ ÎπÑÍµêÌï¥ Î≥¥ÏÑ∏Ïöî
- Î≥∏Ïù∏ Ìè¨Ìä∏Ìè¥Î¶¨Ïò§ Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÎßåÎì† Î™®Îç∏Ïùò Permutation importancesÎ•º Í≥ÑÏÇ∞Ìï¥ Î≥¥ÏÑ∏Ïöî.
- **(Urclass Quiz) breast cancer Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÎßåÎì† Î™®Îç∏Ïùò Permutation importancesÎ•º Í≥ÑÏÇ∞ÌïòÍ≥† Í∞ÄÏû• ÎÜíÏùÄ Í∞íÏùÑ Í∞ÄÏßÄÎäî ÌäπÏÑ±Ïùò Ïù¥Î¶ÑÏùÑ Ï†úÏ∂úÌïòÏÑ∏Ïöî.**
    - eli5 PermutationImportance ÏÇ¨Ïö©
    - scoring = 'accuracy'
    - n_iter = 5
    - random_state = 2

```python
### (Î≥∏Ïù∏ Ìè¨Ìä∏Ìè¥Î¶¨Ïò§ Í≥ºÏ†ú) Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ### 
```

```python
### (Urclass Quiz) Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###

import eli5
from eli5.sklearn import PermutationImportance

permut = PermutationImportance(
    XGBClassifier().fit(X_train, y_train),
    scoring = 'accuracy',
    n_iter = 5,
    random_state = 2
).fit(X_train, y_train)

result = pd.Series(
    permut.feature_importances_,
    index= cancer_dataset.feature_names.tolist(),
    name= 'feature_importance').sort_values()

eli5.show_weights(
    permut, 
    top=None, # top n ÏßÄÏ†ï Í∞ÄÎä•, None Ïùº Í≤ΩÏö∞ Î™®Îì† ÌäπÏÑ± 
    feature_names= cancer_dataset.feature_names # list ÌòïÏãùÏúºÎ°ú ÎÑ£Ïñ¥Ïïº Ìï©ÎãàÎã§
)
```
