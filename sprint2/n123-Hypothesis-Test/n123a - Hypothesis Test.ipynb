# 1. One-Sample T-test 
한 샘플의 평균값과 특정값(모수)을 비교하는 검정 방법입니다.

$H_0$ : $\mu_{sample} = \mu_{population}$

$H_a$ : $\mu_{sample} ≠ \mu_{population}$ 

- 데이터 불러오기 <br>
  [서울시 가로수 데이터셋](https://data.seoul.go.kr/dataList/367/S/2/datasetView.do)을 사용합니다. 다음 링크를 참조하세요. <br>
<https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt>

- 서울시 구별 `이팝나무` 데이터를 모집단으로 하였을 때, 10개의 구를 샘플링한 `이팝나무` 수의 평균과 모집단의 평균을 비교하는 가설검정을 시행하세요.   
  - `서울시`, `공원녹지사업소`, `시설관리공단` 데이터는 제외합니다. 
  - 10개의 구를 중복을 허용하여 샘플링합니다. 
    - seed는 123으로 설정하세요. 
  - scipy 라이브러리의 `.ttest_1samp()`를 사용합니다. <br>
    https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_1samp.html
  - 가설검정 결과의 p-value 값을 `pv1`에 저장하세요. (`.pvalue`를 사용합니다.)

```python
# 이 곳에 답안을 작성하시기 바랍니다.
import pandas as pd
import numpy as np
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt'
tree = pd.read_table(url)

tree
```

```python
tree = tree[tree.자치구 != '서울시'] ##서울시, 공원녹지사업소, 시설관리공단 데이터는 제외합니다. 
tree = tree[tree.자치구 != '공원녹지사업소']
tree = tree[tree.자치구 != '시설관리공단']


a =[]
a = tree.iloc[0]   ### 열 이름 바꿔주기위해 저장해놓는 작업
tree = tree.rename(columns = a)    ## .종별x 말고 나무이름으로 바꿔줌.
tree = tree.drop([tree.index[0]]).reset_index(drop=True)   ### 첫행 삭제 후 인덱스 재정렬
tree
```

```python
import random
tree['이팝나무'] = tree['이팝나무'].str.replace(',','').astype(int)
tree
```

```python
sample10000_means = []
sample10000 = tree['이팝나무'].sample(10000,random_state=123, replace=True)  ## 10000개 데이터 중복추출
sample10000_means.append(sample10000.mean())
sample10000_means
```

```python
pop_means =[] 
pop_means= np.array(pop_means)
pop_means = tree.이팝나무.mean()
pop_means  ## 모집단 평균 = 688.64 , 샘플링 평균 = 622.73 으로 같지않음 
```

```python
from scipy import stats
sample10000_ttest = stats.ttest_1samp(sample10000, pop_means)
pv1 = sample10000_ttest.pvalue
pv1    ## p-value = 0.79로 0.05보다 크므로 귀무가설을 채택
```

# 2. Two-Sample T-test
두 샘플의 평균값을 비교하는 검정 방법입니다. 

$H_0$ : $\mu_1 = \mu_2$

$H_a$ : $\mu_1 ≠ \mu_2$

- 단측 검정(One-tailed Test)을 사용하여 평균 `느티나무` 수와 평균 `왕벚나무` 수를 비교하는 가설검정을 시행하세요. 
  - `서울시`, `공원녹지사업소`, `시설관리공단` 데이터는 제외합니다. 
  - scipy 라이브러리의 `.ttest_ind()`를 사용합니다.

    https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html
  - '평균 느티나무 수가 왕벚나무 수보다 통계적으로 유의미하게 크다'는 가설검정 결과의 pvalue 값을 `pv2`에 저장하세요. 
  - '평균 왕벚나무 수가 느티나무 수보다 통계적으로 유의미하게 크다'는 가설검정 결과의 pvalue 값을 `pv3`에 저장하세요. 
  - 평균 `느티나무` 수와 평균 `왕벚나무` 수를 비교하는 가설검정 결과를 `ans`에 저장하세요. 이때 `ans`는 다음 세 가지 중 하나입니다. 
    - 느티나무 (평균 느티나무 수가 왕벚나무 수보다 통계적으로 유의하게 큰 경우)
    - 왕벚나무 (평균 왕벚나무 수가 느티나무 수보다 통계적으로 유의하게 큰 경우)
    - 같다. (통계적으로 유의미한 차이가 없다.) 

```python
# 1번은 모집단데이터라고 가정, 2번은 샘플데이터로 가정 (디코에 이게 샘플데이터라는 지혜님 설명을 나중에 2번풀때 읽음)
tree['느티나무'] = tree['느티나무'].str.replace(',','').astype(int)
tree['왕벚나무'] = tree['왕벚나무'].str.replace(',','').astype(int)
```

```python
pop2 = tree.느티나무
pop2 = np.array(pop2)
pop3 = tree.왕벚나무
pop3 = np.array(pop3)
print(pop2.var())
print(pop3.var())
print('pop2평균 = ',pop2.mean())
print('pop3평균 = ', pop3.mean())
```

```python
pv2 = stats.ttest_ind(pop2, pop3, alternative='greater',equal_var = False).pvalue
pv3 = stats.ttest_ind(pop3, pop2, alternative='greater',equal_var = False).pvalue
print('pv2 = ',pv2)
print('pv3 = ',pv3)   #pv2 = 0.32  pv3 = 0.68

if pv2 < 0.05 :   ## 귀무가설: 느티나무가 왕벚나무보다 작거나같다 인데  p값이 0.05 이하면 귀무가설 기각후 대립가설채택 = 느티나무가 왕벚나무보다 크다
  ans = '느티나무'
elif pv3 < 0.05 :  ## 귀무가설 : 왕벚나무가 느티나무보다 작거나같다   p값이 0.05 이하면 귀무가설 기각후 대립가설채택 = 왕벚나무가 느티나무보다 크다
  ans = '왕벚나무'
else :            ## 둘다 p가 0.05 이상이면 '느티나무가 왕벚나무보다 작거나같다 and 왕벚나무가 느티나무보다 작거나같다' = 같다 
  ans = '같다'
print(ans)
 
```
