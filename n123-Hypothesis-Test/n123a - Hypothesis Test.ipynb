# 1. One-Sample T-test 
Ìïú ÏÉòÌîåÏùò ÌèâÍ∑†Í∞íÍ≥º ÌäπÏ†ïÍ∞í(Î™®Ïàò)ÏùÑ ÎπÑÍµêÌïòÎäî Í≤ÄÏ†ï Î∞©Î≤ïÏûÖÎãàÎã§.

$H_0$ : $\mu_{sample} = \mu_{population}$

$H_a$ : $\mu_{sample} ‚â† \mu_{population}$ 

- Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ <br>
  [ÏÑúÏö∏Ïãú Í∞ÄÎ°úÏàò Îç∞Ïù¥ÌÑ∞ÏÖã](https://data.seoul.go.kr/dataList/367/S/2/datasetView.do)ÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§. Îã§Ïùå ÎßÅÌÅ¨Î•º Ï∞∏Ï°∞ÌïòÏÑ∏Ïöî. <br>
<https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt>

- ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ `Ïù¥ÌåùÎÇòÎ¨¥` Îç∞Ïù¥ÌÑ∞Î•º Î™®ÏßëÎã®ÏúºÎ°ú ÌïòÏòÄÏùÑ Îïå, 10Í∞úÏùò Íµ¨Î•º ÏÉòÌîåÎßÅÌïú `Ïù¥ÌåùÎÇòÎ¨¥` ÏàòÏùò ÌèâÍ∑†Í≥º Î™®ÏßëÎã®Ïùò ÌèâÍ∑†ÏùÑ ÎπÑÍµêÌïòÎäî Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî.   
  - `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§. 
  - 10Í∞úÏùò Íµ¨Î•º Ï§ëÎ≥µÏùÑ ÌóàÏö©ÌïòÏó¨ ÏÉòÌîåÎßÅÌï©ÎãàÎã§. 
    - seedÎäî 123ÏúºÎ°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî. 
  - scipy ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò `.ttest_1samp()`Î•º ÏÇ¨Ïö©Ìï©ÎãàÎã§. <br>
    https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_1samp.html
  - Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò p-value Í∞íÏùÑ `pv1`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. (`.pvalue`Î•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.)

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§.
import pandas as pd
import numpy as np
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt'
tree = pd.read_table(url)

tree
```

```python
tree = tree[tree.ÏûêÏπòÍµ¨ != 'ÏÑúÏö∏Ïãú'] ##ÏÑúÏö∏Ïãú, Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå, ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã® Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§. 
tree = tree[tree.ÏûêÏπòÍµ¨ != 'Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå']
tree = tree[tree.ÏûêÏπòÍµ¨ != 'ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®']


a =[]
a = tree.iloc[0]   ### Ïó¥ Ïù¥Î¶Ñ Î∞îÍøîÏ£ºÍ∏∞ÏúÑÌï¥ Ï†ÄÏû•Ìï¥ÎÜìÎäî ÏûëÏóÖ
tree = tree.rename(columns = a)    ## .Ï¢ÖÎ≥Ñx ÎßêÍ≥† ÎÇòÎ¨¥Ïù¥Î¶ÑÏúºÎ°ú Î∞îÍøîÏ§å.
tree = tree.drop([tree.index[0]]).reset_index(drop=True)   ### Ï≤´Ìñâ ÏÇ≠Ï†ú ÌõÑ Ïù∏Îç±Ïä§ Ïû¨Ï†ïÎ†¨
tree
```

```python
import random
tree['Ïù¥ÌåùÎÇòÎ¨¥'] = tree['Ïù¥ÌåùÎÇòÎ¨¥'].str.replace(',','').astype(int)
tree
```

```python
sample10000_means = []
sample10000 = tree['Ïù¥ÌåùÎÇòÎ¨¥'].sample(10000,random_state=123, replace=True)  ## 10000Í∞ú Îç∞Ïù¥ÌÑ∞ Ï§ëÎ≥µÏ∂îÏ∂ú
sample10000_means.append(sample10000.mean())
sample10000_means
```

```python
pop_means =[] 
pop_means= np.array(pop_means)
pop_means = tree.Ïù¥ÌåùÎÇòÎ¨¥.mean()
pop_means  ## Î™®ÏßëÎã® ÌèâÍ∑† = 688.64 , ÏÉòÌîåÎßÅ ÌèâÍ∑† = 622.73 ÏúºÎ°ú Í∞ôÏßÄÏïäÏùå 
```

```python
from scipy import stats
sample10000_ttest = stats.ttest_1samp(sample10000, pop_means)
pv1 = sample10000_ttest.pvalue
pv1    ## p-value = 0.79Î°ú 0.05Î≥¥Îã§ ÌÅ¨ÎØÄÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Ï±ÑÌÉù
```

# 2. Two-Sample T-test
Îëê ÏÉòÌîåÏùò ÌèâÍ∑†Í∞íÏùÑ ÎπÑÍµêÌïòÎäî Í≤ÄÏ†ï Î∞©Î≤ïÏûÖÎãàÎã§. 

$H_0$ : $\mu_1 = \mu_2$

$H_a$ : $\mu_1 ‚â† \mu_2$

- Îã®Ï∏° Í≤ÄÏ†ï(One-tailed Test)ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÌèâÍ∑† `ÎäêÌã∞ÎÇòÎ¨¥` ÏàòÏôÄ ÌèâÍ∑† `ÏôïÎ≤öÎÇòÎ¨¥` ÏàòÎ•º ÎπÑÍµêÌïòÎäî Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî. 
  - `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§. 
  - scipy ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò `.ttest_ind()`Î•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.

    https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html
  - 'ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥ ÏàòÍ∞Ä ÏôïÎ≤öÎÇòÎ¨¥ ÏàòÎ≥¥Îã§ ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ÌïòÍ≤å ÌÅ¨Îã§'Îäî Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalue Í∞íÏùÑ `pv2`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. 
  - 'ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ ÏàòÍ∞Ä ÎäêÌã∞ÎÇòÎ¨¥ ÏàòÎ≥¥Îã§ ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ÌïòÍ≤å ÌÅ¨Îã§'Îäî Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalue Í∞íÏùÑ `pv3`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. 
  - ÌèâÍ∑† `ÎäêÌã∞ÎÇòÎ¨¥` ÏàòÏôÄ ÌèâÍ∑† `ÏôïÎ≤öÎÇòÎ¨¥` ÏàòÎ•º ÎπÑÍµêÌïòÎäî Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÎ•º `ans`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. Ïù¥Îïå `ans`Îäî Îã§Ïùå ÏÑ∏ Í∞ÄÏßÄ Ï§ë ÌïòÎÇòÏûÖÎãàÎã§. 
    - ÎäêÌã∞ÎÇòÎ¨¥ (ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥ ÏàòÍ∞Ä ÏôïÎ≤öÎÇòÎ¨¥ ÏàòÎ≥¥Îã§ ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÍ≤å ÌÅ∞ Í≤ΩÏö∞)
    - ÏôïÎ≤öÎÇòÎ¨¥ (ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ ÏàòÍ∞Ä ÎäêÌã∞ÎÇòÎ¨¥ ÏàòÎ≥¥Îã§ ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÍ≤å ÌÅ∞ Í≤ΩÏö∞)
    - Í∞ôÎã§. (ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏Ìïú Ï∞®Ïù¥Í∞Ä ÏóÜÎã§.) 

```python
# 1Î≤àÏùÄ Î™®ÏßëÎã®Îç∞Ïù¥ÌÑ∞ÎùºÍ≥† Í∞ÄÏ†ï, 2Î≤àÏùÄ ÏÉòÌîåÎç∞Ïù¥ÌÑ∞Î°ú Í∞ÄÏ†ï (ÎîîÏΩîÏóê Ïù¥Í≤å ÏÉòÌîåÎç∞Ïù¥ÌÑ∞ÎùºÎäî ÏßÄÌòúÎãò ÏÑ§Î™ÖÏùÑ ÎÇòÏ§ëÏóê 2Î≤àÌíÄÎïå ÏùΩÏùå)
tree['ÎäêÌã∞ÎÇòÎ¨¥'] = tree['ÎäêÌã∞ÎÇòÎ¨¥'].str.replace(',','').astype(int)
tree['ÏôïÎ≤öÎÇòÎ¨¥'] = tree['ÏôïÎ≤öÎÇòÎ¨¥'].str.replace(',','').astype(int)
```

```python
pop2 = tree.ÎäêÌã∞ÎÇòÎ¨¥
pop2 = np.array(pop2)
pop3 = tree.ÏôïÎ≤öÎÇòÎ¨¥
pop3 = np.array(pop3)
print(pop2.var())
print(pop3.var())
print('pop2ÌèâÍ∑† = ',pop2.mean())
print('pop3ÌèâÍ∑† = ', pop3.mean())
```

```python
pv2 = stats.ttest_ind(pop2, pop3, alternative='greater',equal_var = False).pvalue
pv3 = stats.ttest_ind(pop3, pop2, alternative='greater',equal_var = False).pvalue
print('pv2 = ',pv2)
print('pv3 = ',pv3)   #pv2 = 0.32  pv3 = 0.68

if pv2 < 0.05 :   ## Í∑ÄÎ¨¥Í∞ÄÏÑ§: ÎäêÌã∞ÎÇòÎ¨¥Í∞Ä ÏôïÎ≤öÎÇòÎ¨¥Î≥¥Îã§ ÏûëÍ±∞ÎÇòÍ∞ôÎã§ Ïù∏Îç∞  pÍ∞íÏù¥ 0.05 Ïù¥ÌïòÎ©¥ Í∑ÄÎ¨¥Í∞ÄÏÑ§ Í∏∞Í∞ÅÌõÑ ÎåÄÎ¶ΩÍ∞ÄÏÑ§Ï±ÑÌÉù = ÎäêÌã∞ÎÇòÎ¨¥Í∞Ä ÏôïÎ≤öÎÇòÎ¨¥Î≥¥Îã§ ÌÅ¨Îã§
  ans = 'ÎäêÌã∞ÎÇòÎ¨¥'
elif pv3 < 0.05 :  ## Í∑ÄÎ¨¥Í∞ÄÏÑ§ : ÏôïÎ≤öÎÇòÎ¨¥Í∞Ä ÎäêÌã∞ÎÇòÎ¨¥Î≥¥Îã§ ÏûëÍ±∞ÎÇòÍ∞ôÎã§   pÍ∞íÏù¥ 0.05 Ïù¥ÌïòÎ©¥ Í∑ÄÎ¨¥Í∞ÄÏÑ§ Í∏∞Í∞ÅÌõÑ ÎåÄÎ¶ΩÍ∞ÄÏÑ§Ï±ÑÌÉù = ÏôïÎ≤öÎÇòÎ¨¥Í∞Ä ÎäêÌã∞ÎÇòÎ¨¥Î≥¥Îã§ ÌÅ¨Îã§
  ans = 'ÏôïÎ≤öÎÇòÎ¨¥'
else :            ## ÎëòÎã§ pÍ∞Ä 0.05 Ïù¥ÏÉÅÏù¥Î©¥ 'ÎäêÌã∞ÎÇòÎ¨¥Í∞Ä ÏôïÎ≤öÎÇòÎ¨¥Î≥¥Îã§ ÏûëÍ±∞ÎÇòÍ∞ôÎã§ and ÏôïÎ≤öÎÇòÎ¨¥Í∞Ä ÎäêÌã∞ÎÇòÎ¨¥Î≥¥Îã§ ÏûëÍ±∞ÎÇòÍ∞ôÎã§' = Í∞ôÎã§ 
  ans = 'Í∞ôÎã§'
print(ans)
 
```
