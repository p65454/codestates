# 1. customer ÌÖåÏù¥Î∏îÏùò Îç∞Ïù¥ÌÑ∞ Ïù¥Ïäà Ìï¥Í≤∞ÌïòÍ∏∞ 
- customer Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Î∂àÎü¨ÏòµÎãàÎã§. 
- ÎßàÏßÄÎßâ rowÎ•º Ï†úÍ±∞Ìï©ÎãàÎã§. 
- `customer_id`, `store_id`, `address_id`, `active` Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ intÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§. 
- `name`ÏóêÏÑú Ï≤òÏùå Ïù¥Î¶ÑÏùÄ `first_name` Ïª¨ÎüºÏóê, Îã§Ïùå Ïù¥Î¶ÑÏùÄ `last_name` Ïª¨ÎüºÏóê Ï†ÄÏû•Ìï©ÎãàÎã§. Ïù¥ÌõÑ `name` Ïª¨ÎüºÏùÄ Ï†úÍ±∞Ìï©ÎãàÎã§.

````python
import pandas as pd
import numpy as np
from google.colab import files
files.upload()
````

````python
cus = pd.read_csv('customer')
cus.head()
````

````python
cus.tail()
````

````python
cus_clean = cus.copy()
cus_clean = cus_clean.drop([cus_clean.index[-1]])
cus_clean.tail()
````

````python
cus_clean.dtypes  #customer_id, store_id, address_id, active Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ intÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
````

````python
cus_clean['customer_id'] = pd.to_numeric(cus_clean['customer_id'])
cus_clean = cus_clean.astype({'store_id':'int',
                              'address_id':'int',
                              'active':'int'})
cus_clean.dtypes
````

````python
cus_clean.isnull().sum()
````

````python
#nameÏóêÏÑú Ï≤òÏùå Ïù¥Î¶ÑÏùÄ first_name Ïª¨ÎüºÏóê, Îã§Ïùå Ïù¥Î¶ÑÏùÄ last_name Ïª¨ÎüºÏóê Ï†ÄÏû•Ìï©ÎãàÎã§. Ïù¥ÌõÑ name Ïª¨ÎüºÏùÄ Ï†úÍ±∞Ìï©ÎãàÎã§.

first_name, last_name = cus_clean['name'].str.split(',',1).str
cus_clean = cus_clean.drop(['name'],axis = 1)
````

````python
first_name = first_name.to_frame()
last_name = last_name.to_frame()
````

````python
first_name.rename(columns={'name':'firstname'}, inplace = True)
first_name.head()
````

````python
last_name.rename(columns={'name':'lastname'}, inplace = True)
last_name.head()
````

````python
cus_clean.head()
````

````python
cus_clean = pd.concat([cus_clean,first_name,last_name],axis = 1)
cus_clean.head()
````

````python
cus_clean.head(15)
````

````python
#customer_id Ïª¨ÎüºÏóê ÎåÄÌïòÏó¨ Ïò§Î¶ÑÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨Ìï©ÎãàÎã§. Ïù¥ÌõÑ .reset_index()Î•º ÏÇ¨Ïö©ÌïòÏó¨ Ïù∏Îç±Ïä§Î•º Ïû¨ÏÑ§Ï†ï Ìï¥ Ï£ºÏÑ∏Ïöî.
cus_clean = cus_clean.sort_values(by = 'customer_id').reset_index()
cus_clean.head(15)
````
